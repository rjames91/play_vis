```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

import numpy as np
import pylab as plt
import sim_tools.connectors.kernel_connectors as kconn
from spike_tools.vis import my_imshow
from sim_tools.vis import plot_connector_3d
import sim_tools.kernels.center_surround as csgen
import sim_tools.kernels.gabor as gabgen

from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
imgw = 4
imgh = 4
half_imgw = imgw//2
half_imgh = imgh//2

kernel_w = 3
std_dev = 0.8
sd_mult = 6.7

freq = 5.
pi_divs = 7
pi_frac = np.pi/float(pi_divs)
gk = gabgen.multi_gabor(kernel_w, [i*pi_frac for i in range(pi_divs)],
                        std_dev, freq)
csk = csgen.center_surround_kernel(kernel_w, std_dev, sd_mult)

print(csk)
```

```python
colstep=2
rowstep=2
col_start=0
row_start=0
cs_cons = kconn.full_kernel_connector(imgw, imgh, csk, 
                                      col_step=colstep, row_step=rowstep,
                                      col_start=col_start, 
                                      row_start=row_start)
# for c in cs_cons[0]:
#     print(c)

gk_cons = []
for gki in gk:
    gk_cons.append( kconn.full_kernel_connector(imgw, imgh, gki, 
                                                col_step=colstep, 
                                                row_step=rowstep,
                                                col_start=col_start, 
                                                row_start=row_start) )
```

```python
fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
ax = Axes3D(fig)
plot_connector_3d(ax, imgw, imgh, cs_cons, col_start, row_start, 
                  colstep, rowstep)
plt.show()
```

```python
for gk_c in gk_cons:
    fig = plt.figure()
    # ax = fig.add_subplot(111, projection='3d')
    ax = Axes3D(fig)
    plot_connector_3d(ax, imgw, imgh, gk_c, col_start, row_start, 
                      colstep, rowstep)
    plt.show()
```
